```python
import pandas as pd
import matplotlib.pyplot as plt

# Passo 1: Importar os dados do arquivo CSV corrigido
df = pd.read_csv('./data.csv')

# Filtrar as colunas necessárias
df = df[['Country Name', '2000 [YR2000]', '2023 [YR2023]']]

# Converter as colunas de acesso à eletricidade para tipo numérico
df['2000 [YR2000]'] = pd.to_numeric(df['2000 [YR2000]'], errors='coerce')
df['2023 [YR2023]'] = pd.to_numeric(df['2023 [YR2023]'], errors='coerce')

# Passo 2: Explicar em Markdown o que está sendo analisado (convertido para comentário no código)
"""
# Análise de Acesso à Eletricidade

Estamos analisando o percentual da população com acesso à eletricidade em três países (Brasil, Estados Unidos e China) nos anos 2000 e 2023.

### Dados Fornecidos

- **Country Name**: Nome do país.
- **2000 [YR2000]**: Percentual da população com acesso à eletricidade no ano 2000.
- **2023 [YR2023]**: Percentual da população com acesso à eletricidade no ano 2023.

### Objetivo da Análise

Comparar o acesso à eletricidade em 2000 e 2023.

### Conclusão

Os gráficos mostram a evolução do acesso à eletricidade de 2000 a 2023.
"""

# Configurar os dados para visualização
countries = df['Country Name']
access_2000 = df['2000 [YR2000]']
access_2023 = df['2023 [YR2023]']

# Gráfico de Linha
plt.figure(figsize=(10, 6))
plt.plot(countries, access_2000, marker='o', label='2000')
plt.plot(countries, access_2023, marker='o', label='2023')
plt.title('Acesso à Eletricidade (% da População) em 2000 e 2023')
plt.xlabel('País')
plt.ylabel('Acesso à Eletricidade (%)')
plt.ylim(90, 105)  # Ajustar limite do eixo y para melhor visualização
plt.legend()
plt.grid(True)
plt.show()

# Gráfico de Barra
plt.figure(figsize=(10, 6))
plt.bar(countries, access_2000, color='skyblue', label='2000', width=0.4, align='center')
plt.bar(countries, access_2023, color='orange', label='2023', width=0.4, align='edge')
plt.title('Acesso à Eletricidade (% da População) em 2000 e 2023')
plt.xlabel('País')
plt.ylabel('Acesso à Eletricidade (%)')
plt.ylim(90, 105)  # Ajustar limite do eixo y para melhor visualização
plt.legend()
plt.show()

# Gráfico de Barras Agrupadas
width = 0.35
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(countries, access_2000, width, label='2000', color='skyblue')
ax.bar(countries, access_2023, width, bottom=access_2000, label='2023', color='orange')
ax.set_title('Acesso à Eletricidade (% da População) em 2000 e 2023')
ax.set_xlabel('País')
ax.set_ylabel('Acesso à Eletricidade (%)')
ax.legend()

plt.show()

# Histograma
plt.figure(figsize=(10, 6))
plt.hist([access_2000, access_2023], bins=3, color=['green', 'red'], edgecolor='black', label=['2000', '2023'])
plt.title('Distribuição do Acesso à Eletricidade em 2000 e 2023')
plt.xlabel('Acesso à Eletricidade (%)')
plt.ylabel('Frequência')
plt.legend()
plt.show()

# Gráfico de Pizza (para 2023)
plt.figure(figsize=(10, 6))
plt.pie(access_2023, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Proporção de Acesso à Eletricidade por País em 2023')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Gráfico Scatter
plt.figure(figsize=(10, 6))
plt.scatter(countries, access_2000, color='blue', label='2000')
plt.scatter(countries, access_2023, color='red', label='2023')
plt.title('Acesso à Eletricidade (% da População) em 2000 e 2023')
plt.xlabel('País')
plt.ylabel('Acesso à Eletricidade (%)')
plt.ylim(90, 105)  # Ajustar limite do eixo y para melhor visualização
plt.legend()
plt.grid(True)
plt.show()
```
