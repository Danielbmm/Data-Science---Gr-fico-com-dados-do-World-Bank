```python
import pandas as pd
import matplotlib.pyplot as plt

# Passo 1: Importar os dados do arquivo CSV corrigido
df = pd.read_csv('./data.csv')

# Filtrar as colunas necessrias
df = df[['Country Name', '2000 [YR2000]', '2023 [YR2023]']]

# Converter as colunas de acesso  eletricidade para tipo numrico
df['2000 [YR2000]'] = pd.to_numeric(df['2000 [YR2000]'], errors='coerce')
df['2023 [YR2023]'] = pd.to_numeric(df['2023 [YR2023]'], errors='coerce')

# Passo 2: Explicar em Markdown o que est sendo analisado (convertido para comentrio no cdigo)
"""
# Anlise de Acesso  Eletricidade

Estamos analisando o percentual da populao com acesso  eletricidade em trs pases (Brasil, Estados Unidos e China) nos anos 2000 e 2023.

### Dados Fornecidos

- **Country Name**: Nome do pas.
- **2000 [YR2000]**: Percentual da populao com acesso  eletricidade no ano 2000.
- **2023 [YR2023]**: Percentual da populao com acesso  eletricidade no ano 2023.

### Objetivo da Anlise

Comparar o acesso  eletricidade em 2000 e 2023.

### Concluso

Os grficos mostram a evoluo do acesso  eletricidade de 2000 a 2023.
"""

# Configurar os dados para visualizao
countries = df['Country Name']
access_2000 = df['2000 [YR2000]']
access_2023 = df['2023 [YR2023]']

# Grfico de Linha
plt.figure(figsize=(10, 6))
plt.plot(countries, access_2000, marker='o', label='2000')
plt.plot(countries, access_2023, marker='o', label='2023')
plt.title('Acesso  Eletricidade (% da Populao) em 2000 e 2023')
plt.xlabel('Pas')
plt.ylabel('Acesso  Eletricidade (%)')
plt.ylim(90, 105)  # Ajustar limite do eixo y para melhor visualizao
plt.legend()
plt.grid(True)
plt.show()

# Grfico de Barra
plt.figure(figsize=(10, 6))
plt.bar(countries, access_2000, color='skyblue', label='2000', width=0.4, align='center')
plt.bar(countries, access_2023, color='orange', label='2023', width=0.4, align='edge')
plt.title('Acesso  Eletricidade (% da Populao) em 2000 e 2023')
plt.xlabel('Pas')
plt.ylabel('Acesso  Eletricidade (%)')
plt.ylim(90, 105)  # Ajustar limite do eixo y para melhor visualizao
plt.legend()
plt.show()

# Grfico de Barras Agrupadas
width = 0.35
fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(countries, access_2000, width, label='2000', color='skyblue')
ax.bar(countries, access_2023, width, bottom=access_2000, label='2023', color='orange')
ax.set_title('Acesso  Eletricidade (% da Populao) em 2000 e 2023')
ax.set_xlabel('Pas')
ax.set_ylabel('Acesso  Eletricidade (%)')
ax.legend()

plt.show()

# Histograma
plt.figure(figsize=(10, 6))
plt.hist([access_2000, access_2023], bins=3, color=['green', 'red'], edgecolor='black', label=['2000', '2023'])
plt.title('Distribuio do Acesso  Eletricidade em 2000 e 2023')
plt.xlabel('Acesso  Eletricidade (%)')
plt.ylabel('Frequncia')
plt.legend()
plt.show()

# Grfico de Pizza (para 2023)
plt.figure(figsize=(10, 6))
plt.pie(access_2023, labels=countries, autopct='%1.1f%%', startangle=140)
plt.title('Proporo de Acesso  Eletricidade por Pas em 2023')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle.
plt.show()

# Grfico Scatter
plt.figure(figsize=(10, 6))
plt.scatter(countries, access_2000, color='blue', label='2000')
plt.scatter(countries, access_2023, color='red', label='2023')
plt.title('Acesso  Eletricidade (% da Populao) em 2000 e 2023')
plt.xlabel('Pas')
plt.ylabel('Acesso  Eletricidade (%)')
plt.ylim(90, 105)  # Ajustar limite do eixo y para melhor visualizao
plt.legend()
plt.grid(True)
plt.show()
```
